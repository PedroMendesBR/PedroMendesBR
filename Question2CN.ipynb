```python
#Código questão 2

import numpy as np
import matplotlib.pyplot as plt

# Definir a função
def f(x):
    return np.exp(-x**2) - np.cos(x)

# Definir intervalo e tolerância
a = 1
b = 2
epsilon = 1e-4

# Método da bissecção
c = None
while (b - a) > epsilon:
    c = (a + b) / 2
    if f(c) == 0:
        break
    elif f(c) * f(a) > 0:
        a = c
    else:
        b = c
bisseccao_solution = c

# Método de Newton-Raphson
x0 = 1.5
while abs(f(x0)) > epsilon:
    x0 = x0 - f(x0) / (np.exp(-x0**2) + np.sin(x0))
newton_solution = x0

# Método da posição falsa
x1 = a
x2 = b
x_posicao_falsa = 1
while abs(f(x_posicao_falsa)) > epsilon:
    x_posicao_falsa = (x1 * f(x2) - x2 * f(x1)) / (f(x2) - f(x1))
    if f(x_posicao_falsa) * f(x1) > 0:
        x1 = x_posicao_falsa
    else:
        x2 = x_posicao_falsa
posicao_falsa_solution = x_posicao_falsa

# Método da secante
iteracoes_secante = 0
def secante(f, x0, x1, epsilon):
    global iteracoes_secante
    x = 0
    while abs(f(x1)) > epsilon:
        x = x1 - (f(x1) * (x1 - x0)) / (f(x1) - f(x0))
        x0 = x1
        x1 = x
        iteracoes_secante += 1
    return x

secante_solution = secante(f, a, b, epsilon)

# Imprimir as soluções
print("Solução pelo método da bissecção:", bisseccao_solution)
print("Solução pelo método de Newton-Raphson:", newton_solution)
print("Solução pelo método da posição falsa:", posicao_falsa_solution)
print("Solução pelo método da secante:", secante_solution, "iterações:", iteracoes_secante)

# Gerar pontos para o gráfico
x = np.linspace(a, b, 1000)
y = f(x)  # Agora f(x) funciona corretamente com arrays

# Plotar o gráfico
plt.plot(x, y, label="f(x)")
plt.scatter(bisseccao_solution, f(bisseccao_solution), color="red", label="Bissecção", zorder=5)
plt.scatter(newton_solution, f(newton_solution), color="green", label="Newton-Raphson", zorder=5)
plt.scatter(posicao_falsa_solution, f(posicao_falsa_solution), color="blue", label="Posição Falsa", zorder=5)
plt.axhline(0, color="black",linewidth=0.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Métodos Numéricos")
plt.show()

# Comparação da rapidez de convergência
iterations_bisseccao = np.log2((b - a) / epsilon)
iterations_newton = 1
x_newton = x0
while abs(f(x_newton)) > epsilon:
    x_newton = x_newton - f(x_newton) / (np.exp(-x_newton**2) + np.sin(x_newton))
    iterations_newton += 1

iterations_posicao_falsa = 1
x_posicao_falsa = posicao_falsa_solution
x1, x2 = a, b
while abs(f(x_posicao_falsa)) > epsilon:
    x_posicao_falsa = (x1 * f(x2) - x2 * f(x1)) / (f(x2) - f(x1))
    if f(x_posicao_falsa) * f(x1) > 0:
        x1 = x_posicao_falsa
    else:
        x2 = x_posicao_falsa
    iterations_posicao_falsa += 1

print("Rapidez de convergência pelo método da bissecção:", iterations_bisseccao, "iterações")
print("Rapidez de convergência pelo método de Newton-Raphson:", iterations_newton, "iterações")
print("Rapidez de convergência pelo método da posição falsa:", iterations_posicao_falsa, "iterações")
print("Rapidez de convergência pelo método da secante:", iteracoes_secante, "iterações")
```

```python
#Código questão 1 - Apenas rodar. Todas as funções estão definidas dentro do próprio escopo

import numpy as np
import matplotlib.pyplot as plt

# Definir q(x) e p(x)
def q(x):
    return 816*x**3 - 3835*x**2 + 6000*x - 3125

def p(x, w=4):
    return q(x + w)

# Derivada de q(x)
def q_prime(x):
    return 2448*x**2 - 7670*x + 6000

# Derivada de p(x)
def p_prime(x, w=4):
    return q_prime(x + w)

# Função para plotar
def plot_progress(xs, method_name, iteration):
    x_vals = np.linspace(-6, 2, 400)
    y_vals = p(x_vals)
    plt.plot(x_vals, y_vals, label="p(x)")
    plt.scatter(xs, p(np.array(xs)), color="red")
    plt.title(f"{method_name} - Iteração {iteration}")
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(True)
    plt.show()

# Método de Newton-Raphson
def newton_raphson(x0, epsilon=1e-4, max_iter=100):
    x = x0
    iter_count = 0
    x_vals = [x]
    print(f"Newton-Raphson - Iteração {iter_count}: x = {x}, p(x) = {p(x)}")
    plot_progress(x_vals, "Newton-Raphson", iter_count)

    while abs(p(x)) > epsilon and iter_count < max_iter:
        x = x - p(x) / p_prime(x)
        iter_count += 1
        x_vals.append(x)
        print(f"Newton-Raphson - Iteração {iter_count}: x = {x}, p(x) = {p(x)}")
        plot_progress(x_vals, "Newton-Raphson", iter_count)
    return x, iter_count

# Método da Secante
def secante(x0, x1, epsilon=1e-4, max_iter=100):
    iter_count = 0
    x_vals = [x0, x1]
    print(f"Secante - Iteração {iter_count}: x0 = {x0}, x1 = {x1}, p(x1) = {p(x1)}")
    plot_progress(x_vals, "Secante", iter_count)

    while abs(p(x1)) > epsilon and iter_count < max_iter:
        x2 = x1 - (p(x1) * (x1 - x0)) / (p(x1) - p(x0))
        x0, x1 = x1, x2
        iter_count += 1
        x_vals.append(x1)
        print(f"Secante - Iteração {iter_count}: x0 = {x0}, x1 = {x1}, p(x1) = {p(x1)}")
        plot_progress(x_vals, "Secante", iter_count)
    return x1, iter_count

# Método da Bissecção
def bisseccao(a, b, epsilon=1e-4, max_iter=100):
    iter_count = 0
    x_vals = [a, b]
    print(f"Bissecção - Iteração {iter_count}: a = {a}, b = {b}, p(c) = {p((a + b)/2)}")
    plot_progress(x_vals, "Bissecção", iter_count)

    while abs(b - a) > epsilon and iter_count < max_iter:
        c = (a + b) / 2
        if p(c) == 0:
            return c, iter_count
        elif p(a) * p(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1
        x_vals.append(c)
        print(f"Bissecção - Iteração {iter_count}: a = {a}, b = {b}, c = {c}, p(c) = {p(c)}")
        plot_progress(x_vals, "Bissecção", iter_count)
    return (a + b) / 2, iter_count

# Testar os métodos com visualização

# Newton-Raphson
x0_newton = 1.5 - 4
root_newton, iterations_newton = newton_raphson(x0_newton)

# Secante
x0_secante = -4
x1_secante = 1 - 4
root_secante, iterations_secante = secante(x0_secante, x1_secante)

# Bissecção
a_bisseccao = 1 - 4
b_bisseccao = 2 - 4
root_bisseccao, iterations_bisseccao = bisseccao(a_bisseccao, b_bisseccao)
```
